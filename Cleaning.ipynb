```python
import pandas as pd
data=pd.read_csv("/content/ecommerce-sales-dataset/customer_details.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
print("Column Names:", data.columns.tolist())
```

```python
data.isnull().sum()
```

```python
for col in data.select_dtypes(include='object').columns:
    print(f"\nUnique values in {col}:")
    print(data[col].unique()[:10])
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)  # OR
data.fillna(method='ffill', inplace=True)
```

```python
#Check for and remove duplicates
print("Duplicates:", data.duplicated().sum())
data.drop_duplicates(inplace=True)
```

```python
print(data.columns)
```

```python
# Check for missing values
print("Missing values:\n", data.isnull().sum())
```

```python
#Standardize categorical text
data['sex'] = data['sex'].str.strip().str.title()
```

```python
data.to_csv("/content/ecommerce-sales-dataset/customer_details.csv", index=False)
```

```python
# Remove unrealistic ages (e.g., negative or > 100)
data = data[(data['customer_age'] > 0) & (data['customer_age'] <= 100)]
```

```python
# Fill missing age with median
data['customer_age'].fillna(data['customer_age'].median(), inplace=True)
```

```python
# Remove negative tenure values (if any)
data = data[data['tenure'] >= 0]
```

```python
# Optional: Fill missing tenure with 0 or median
data['tenure'].fillna(0, inplace=True)
```

```python
data.to_csv('/content/cleaned_data.csv', index=False)
```
