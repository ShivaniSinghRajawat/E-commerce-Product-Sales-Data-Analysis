```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/cleaned_data.csv')
```

```python
print("Dataset Info:")
print(data.info())
```

```python
print("Summary Statistics:")
print(data.describe())
```

```python
print("Null Values:")
print(data.isnull().sum())
```

```python
sns.countplot(x='sex', data=data)
plt.title('Gender Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
sns.histplot(data['customer_age'], bins=10, kde=True)
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(data['tenure'], bins=10, kde=True)
plt.title('Tenure Distribution')
plt.xlabel('Tenure (Years)')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.scatterplot(x='customer_age', y='tenure', data=data)
plt.title('Customer Age vs Tenure')
plt.xlabel('Age')
plt.ylabel('Tenure')
plt.show()
```
