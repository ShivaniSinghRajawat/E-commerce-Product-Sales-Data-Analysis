```python
import pandas as pd
```

```python
data = pd.read_csv('basket_details.csv')
```

```python
print("Preview of data:\n", data.head())
```

```python
print("\nColumn names:", data.columns.tolist())
```

```python
print("\nData types:\n", data.dtypes)
```

```python
print("\nMissing values:\n", data.isnull().sum())
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.columns = data.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
if 'product' in data.columns:
    data['product'] = data['product'].astype(str).str.strip().str.title()

if 'category' in data.columns:
    data['category'] = data['category'].astype(str).str.strip().str.title()

# Step 6: Fix data types
if 'date' in data.columns:
    data['date'] = pd.to_datetime(data['date'], errors='coerce')

for col in ['price', 'total', 'amount']:
    if col in data.columns:
        data[col] = pd.to_numeric(data[col], errors='coerce')
```

```python
if 'quantity' in data.columns:
    # Remove rows with negative or zero quantity
    data = data[data['quantity'] > 0]

if 'price' in data.columns:
    data = data[data['price'] >= 0]
```

```python
data.fillna({
    'price': data['price'].median() if 'price' in data.columns else None,
    'quantity': 1,
}, inplace=True)
```

```python
print("\nMissing values after cleaning:\n", data.isnull().sum())
```

```python
data.to_csv('/content/cleaned_basket_data.csv', index=False)
print("\n✅ Cleaned data saved to: /content/cleaned_basket_data.csv")
```
